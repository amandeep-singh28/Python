```python
import mysql.connector as connector
print("MySQL conncetor imported successfully")
```

```python
class DatabaseConnection:
    def __init__(self):
        self.con = connector.connect(
            host = "localhost",
            user = "root",
            password = "aman",
            database = "college"
        )
        
        query = """
        CREATE TABLE IF NOT EXISTS student (
            id INT PRIMARY KEY,
            name VARCHAR(20),
            age INT
        )
        """
        cur = self.con.cursor()
        cur.execute(query)
        print("Table 'student' ensured to exist")

    def insert_value(self):
        try:
            query = """
            INSERT INTO student (id, name, age)
            VALUES (2, "Divya", 22)
            """
            cur = self.con.cursor()
            # cur.execute(query, (id, name, age))
            cur.execute(query)
            self.con.commit()
            print("Student inserted successfully")
        except Exception:
            print("Record Already present.")
    
    def update_values(self):
        query = """
        UPDATE student SET name = "Rajkumar"
        WHERE id = 2
        """
        curr = self.con.cursor()
        curr.execute(query)
        self.con.commit()
        print("Record Updated Successfully")

    def fetch_data(self):
        query = """
        SELECT * FROM student
        """
        curr = self.con.cursor()
        curr.execute(query)
        
        rows = curr.fetchall()
        for row in rows:
            print(row)
    
    def delete_data(self):
        query = """
        DELETE FROM student
        WHERE id = 2
        """
        curr = self.con.cursor()
        curr.execute(query)
        print("Deleted Successfully")

db = DatabaseConnection()
db.insert_value()
db.update_values()
db.fetch_data()
db.delete_data()
db.fetch_data()
```
