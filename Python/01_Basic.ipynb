```python
# Use this method(if you want to take input as bool)
value = input("Enter True or False").lower()


if value == "true":
    flag = True
    print(flag)
    print(type(flag))
elif value == "false":
    flag = False
    print(flag)
    print(type(flag))
else:
    print("Invalid Input")
```

```python
# Type Conversion(Type Casting)
# (i) Implicit Type Conversion (automatic)
# (ii) Explicit Type Conversion (manual)

a = 10      # int
b = 2.5     # float

c = a + b
print(c)
print(type(c))
```
