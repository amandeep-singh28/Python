```python
question_list = []
option_list = []
correct_answer = ['a', 'b']
n = 2
count = 0

for i in range(1, n + 1):
    question = input(f"Enter question{i}:")
    question_list.append(question)
    for j in range(0, 1):
        option_a = input(f"Enter option a for question {i}:")
        option_b = input(f"Enter option b for question {i}:")
        option_c = input(f"Enter option c for question {i}:")
        option_d = input(f"Enter option d for question {i}:")
        option_list.append([option_a, option_b, option_c, option_d])

        answer = input(f"Enter your option for question{i}:")
        if (answer == correct_answer[i - 1]):
            count += 1


print(question_list)
print(option_list)
print(count)
```

```python
student = {
    'name' : 'Rahul',
    'marks' : [80, 90, 95]
}
another_student = student.copy() # Shallow Copy
student['marks'].append(99)
another_student['marks'].append(100)
print(student)
print(another_student)
# Shallow copy works as reference
```

```python
import copy
student = {
    'name' : 'Rahul',
    'marks' : [80, 90, 95]
}
another_student = copy.deepcopy(student) # Deep Copy
student['marks'].append(99)
print(student)
print(another_student)
# Deep Copy works as a normal copy
```

```python
def addElementList(list):
    sum = 0
    for i in range(len(list)):
        sum = sum + list[i]
    return sum

def sumOfEvenNumber(list):
    sum = 0
    for i in range(len(list)):
        if (list[i] % 2 == 0):
            sum = sum + list[i]
    return sum    

list = [10, 21, 30, 40, 50]
print(addElementList(list))
print(sumOfEvenNumber(list))
```

```python
def sum_of_nums(*nums):
    total = 0
    for i in nums:
        total += i
    return total
print(sum_of_nums(1, 2, 3, 4, 5))
```

```python
def sum_of_nums(msg,*nums):
    print(msg)
    total = 0
    for i in nums:
        total += i
    return total
print(sum_of_nums(("I am calculating sum: "),1, 2, 3, 4, 5))
```

#### __*kwargs__

```python
def build_profile(name, city, country, hobbies, profession):
    print("Name", name)
    print("City", city)
    print("Country", country)
    print("Hobbies", hobbies)
    print("Profession", profession)
build_profile(name = "Amandeep", city = "Haridwar", country = "India", hobbies = "Cricket", profession = "Student")
```

```python
def build_profile(**things): # Key value pair
    print("Profiles")
    print(things)
    for i, j in things.items():
        print(i, j)
build_profile(name = "Amandeep", city = "Haridwar", country = "India", hobbies = "Cricket", profession = "Student")
```
