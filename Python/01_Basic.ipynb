```python
num1 = int(input("Enter first number:"))
num2 = int(input("Enter second number:"))

print("Addition:", num1 + num2)
print("Multiplication:", num1 * num2)

if (num1 > num2):
    print("Subtraction:", num1 - num2)
    print("Division:", num1 / num2)
else:
    print("Subtraction:", num2 - num1)
    print("Division:", num2 / num1)

```

```python
num1 = int(input("Enter first number:"))
num2 = int(input("Enter second number:"))
choice = int(input("Enter your choice:\n1-Addition\n2-Subtraction\n3-Division\n4-Multiplication"))
match choice:
    case 1:
        print("Addition for two number is:", num1 + num2)
    case 2:
        if(num1 >= num2):
            print("Subtration for two number is:", num1 - num2)
        else:
            print("Subtration for two number is:", num2 - num1)
    case 3:
        print("Division for two number is:", num1 / num2)
    case 4:
        print("Multiplication for two number is:", num1 * num2)
    case _:
        print("Invalid Choice")
```

```python
time = input("Enter time in 24 hr format(HH:MM):")
hour = int(time.split(":")[0])

if 0 <= hour < 12:
    print("Good Morning")
elif 12 <= hour < 17:
    print("Good Afternoon")
elif 17 <= hour < 21:
    print("Good Evening")
else:
    print("Good Night")
```
