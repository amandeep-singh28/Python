```python
num1 = int(input("Enter first number:"))
num2 = int(input("Enter second number:"))
choice = int(input("Enter your choice:\n1-Addition\n2-Subtraction\n3-Division\n4-Multiplication"))
match choice:
    case 1:
        print("Addition for two number is:", num1 + num2)
    case 2:
        if(num1 >= num2):
            print("Subtration for two number is:", num1 - num2)
        else:
            print("Subtration for two number is:", num2 - num1)
    case 3:
        print("Division for two number is:", num1 / num2)
    case 4:
        print("Multiplication for two number is:", num1 * num2)
    case _:
        print("Invalid Choice")
```

```python
question_list = []
option_list = []
correct_answer = ['a', 'b']
n = 2
count = 0

for i in range(1, n + 1):
    question = input(f"Enter question{i}:")
    question_list.append(question)
    for j in range(0, 1):
        option_a = input(f"Enter option a for question {i}:")
        option_b = input(f"Enter option b for question {i}:")
        option_c = input(f"Enter option c for question {i}:")
        option_d = input(f"Enter option d for question {i}:")
        option_list.append([option_a, option_b, option_c, option_d])

        answer = input(f"Enter your option for question{i}:")
        if (answer == correct_answer[i - 1]):
            count += 1


print(question_list)
print(option_list)
print(count)
```

```python
question_list = ["What is your name?", "What is your age?"]
option_list = [['aman', 'amandeep', 'raj', 'ravi'], ['21', '22', '23', '24']]
correct_ans = [1, 2]
user_answer = []
score = 0

for i in range(len(question_list)):
    print(f"Question {i + 1}:")
    print(question_list[i])
    print(f"Choose the correct option(1, 2, 3, 4):")

   
    for k in range(4):
        print(f"{k + 1} - {option_list[i][k]}")

    answer = int(input("Enter your option(1, 2, 3, 4):"))
    if (answer == correct_ans[i]):
        user_answer.append(answer)
        score += 1

print(f"Total Score:{score}")
```
