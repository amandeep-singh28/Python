```python
class Calculator:
    def __init__(self, num1, num2):
        self.num1 = num1
        self.num2 = num2
    def addition(self):
        print(f"Addition of {self.num1} and {self.num2}: {self.num1 + self.num2}")

    def subtraction(self):
        print(f"Subtraction of {self.num1} and {self.num2}: {self.num1 - self.num2}")

    def multiply(self):
        print(f"Multiplication of {self.num1} and {self.num2}: {self.num1 * self.num2}")

    def division(self):
        if (self.num2 == 0):
            print("Zero division Error")
        else:
            print(f"Division of {self.num1} and {self.num2}: {self.num1 / self.num2}")
            
calc1 = Calculator(10, 20)
calc1.addition()
calc1.subtraction()
calc1.multiply()
calc1.division()
```

```python
# employee, empid, name and salary
# method to calculate the annual salary
class Employee:
    def __init__(self, empid, name, salary, day):
        self.empid = empid
        self.name = name
        self.salary = salary
        self.day = day
    def annual_sal(self):
        print(f"{self.name} has annual salary of {self.salary * 12}")
    def month_sal(self):
        per_day_sal = self.salary / 30
        print(f"Monthly salary according to {self.day} working days {per_day_sal * self.day}")
    def check_bonus(self):
        if (self.salary > 30000):
            print("Eligible for bonus")
        else:
            print("Not Eligible")

days = int(input('Enter number of days:'))
emp1 = Employee(101, "Amandeep", 90000, days)
emp1.annual_sal()
emp1.month_sal()
emp1.check_bonus()
```

```python
class Student1:
    def __init__(self, name, grade, percentage):
        self.name = name
        self.grade = grade
        self.percentage = percentage
    
class Student2(Student1):
    def __init__(self, name, grade, percentage, stream):
        super().__init__(name, grade, percentage)
        self.stream = stream
    def display(self):
        print(f"{self.name} belongs to {self.stream} stream has grade {self.grade} and percentage {self.percentage}")

s1 = Student2("Amandeep", "A", 90.6, "CSE")
s1.display()
```

```python
class Shape:
    def area(self):
        print("Area not defined")
class Rectangle(Shape):
    def area(self):
        print("Area of Rectangle")
class Triangle(Shape):
    def area(self):
        print("Area of Triangle")

obj = [Rectangle(), Triangle()]
for s in obj:
    s.area()
```

```python
class Payment:
    def pay(self, amount):
        print("Processing payment...")

class CreditCard(Payment):
    def pay(self, amount):
        print(f"Paid ₹{amount} using Credit Card")

class UPI(Payment):
    def pay(self, amount):
        print(f"Paid ₹{amount} using UPI")

class NetBanking(Payment):
    def pay(self, amount):
        print(f"Paid ₹{amount} using Net Banking")


# Polymorphism in action
payments = [
    CreditCard(),
    UPI(),
    NetBanking()
]

for payment in payments:
    payment.pay(1000)
```
