```python
# Create class car with object car1
class Car:
    def __init__(self, carName, price, fuel):
        self.carName = carName
        self.price = price
        self.fuel = fuel
    def car_details(self):
        print(f"{self.carName}")
car1 = Car("Brezza", 950000, "Petrol")
car1.car_details()
```

```python
class Calculator:
    def __init__(self, num1, num2):
        self.num1 = num1
        self.num2 = num2
    def addition(self):
        print(f"Addition of {self.num1} and {self.num2}: {self.num1 + self.num2}")

    def subtraction(self):
        print(f"Subtraction of {self.num1} and {self.num2}: {self.num1 - self.num2}")

    def multiply(self):
        print(f"Multiplication of {self.num1} and {self.num2}: {self.num1 * self.num2}")

    def division(self):
        if (self.num2 == 0):
            print("Zero division Error")
        else:
            print(f"Division of {self.num1} and {self.num2}: {self.num1 / self.num2}")
            
calc1 = Calculator(10, 20)
calc1.addition()
calc1.subtraction()
calc1.multiply()
calc1.division()
```

```python
# employee, empid, name and salary
# method to calculate the annual salary
class Employee:
    def __init__(self, empid, name, salary, day):
        self.empid = empid
        self.name = name
        self.salary = salary
        self.day = day
    def annual_sal(self):
        print(f"{self.name} has annual salary of {self.salary * 12}")
    def month_sal(self):
        per_day_sal = self.salary / 30
        print(f"Monthly salary according to {self.day} working days {per_day_sal * self.day}")
    def check_bonus(self):
        if (self.salary > 30000):
            print("Eligible for bonus")
        else:
            print("Not Eligible")

days = int(input('Enter number of days:'))
emp1 = Employee(101, "Amandeep", 90000, days)
emp1.annual_sal()
emp1.month_sal()
emp1.check_bonus()
```
