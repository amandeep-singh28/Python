```python
class Student:
    def __init__(self, name, grade, percentage):
        self.name = name
        self.grade = grade
        self.percentage = percentage
student1 = Student("Amandeep", 10, 20)
print(student1.name)
print(student1.grade)
```

```python
# Create class car with object car1
class Car:
    def __init__(self, carName, price, fuel):
        self.carName = carName
        self.price = price
        self.fuel = fuel
    def car_details(self):
        print(f"{self.carName}")
car1 = Car("Brezza", 950000, "Petrol")
car1.car_details()
```

```python
class BankAccount:
    def __init__(self, balance):
        self.balance = balance

    def bank_details(self):
        print(f"{self.balance}")

obj1 = BankAccount(8.98)
print(f"{obj1.balance}")
obj2 = BankAccount(100)
print(obj2.balance)
print(obj1.__dict__)
print(obj2.__dict__)
```

```python
class Calculator:
    def __init__(self, num1, num2):
        self.num1 = num1
        self.num2 = num2
    def addition(self):
        print(f"Addition of {self.num1} and {self.num2}: {self.num1 + self.num2}")

    def subtraction(self):
        print(f"Subtraction of {self.num1} and {self.num2}: {self.num1 - self.num2}")

    def multiply(self):
        print(f"Multiplication of {self.num1} and {self.num2}: {self.num1 * self.num2}")

    def division(self):
        if (self.num2 == 0):
            print("Zero division Error")
        else:
            print(f"Division of {self.num1} and {self.num2}: {self.num1 / self.num2}")
            
calc1 = Calculator(10, 20)
calc1.addition()
calc1.subtraction()
calc1.multiply()
calc1.division()
```
