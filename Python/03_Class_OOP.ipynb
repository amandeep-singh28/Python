```python
class Student:
    def __init__(self, name, grade, percentage):
        self.name = name
        self.grade = grade
        self.percentage = percentage
student1 = Student("Amandeep", 10, 20)
print(student1.name)
print(student1.grade)
```

```python
# Create class car with object car1
class Car:
    def __init__(self, carName, price, fuel):
        self.carName = carName
        self.price = price
        self.fuel = fuel
    def car_details(self):
        print(f"{self.carName}")
car1 = Car("Brezza", 950000, "Petrol")
car1.car_details()
```

```python
class BankAccount:
    def __init__(self, balance):
        self.balance = balance
    def bank_details(self):
        print(f"{self.balance}")
obj1 = BankAccount(8.98)
obj1.bank_details()
```
